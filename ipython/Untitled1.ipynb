```python
import pandas as pd
import matplotlib.pylab as plt
import numpy as np
import matplotlib
```

```python
def read_data(in_file):
    x = pd.read_csv(in_file)
    output_map = {'S125':x['S125'],
                  'MeanEnergyLoss':x['MeanEnergyLoss'],
                 'HE_stoch_standard':x['HE_stoch_standard'],
                 'HE_stoch_strong':x['HE_stoch_strong']}
    return output_map
```

```python
map1 = read_data('outputFe.csv')
map2 = read_data('outputHe.csv')
map3 = read_data('outputO.csv')
map4 = read_data('outputP.csv')
```

```python
def plot_data(in_map,plot_name):
    fig, (ax1,ax2,ax3) = plt.subplots(nrows=3,figsize=(15,10))
    img1= ax1.hist2d(in_map['S125'],in_map['MeanEnergyLoss'],bins=100,cmap=plt.cm.jet,norm=matplotlib.colors.LogNorm())
    ax1.set_xlabel('S125')
    ax1.set_ylabel('MeanEnergyLoss')
    fig.colorbar(img1[3], ax=ax1)
    ax1.set_aspect('auto')
    img2 = ax2.hist2d(in_map['S125'],in_map['HE_stoch_standard'],bins=100,cmap=plt.cm.jet,norm=matplotlib.colors.LogNorm())
    ax2.set_xlabel('S125')
    ax3.set_ylabel('HE_stoch_standard')
    fig.colorbar(img2[3], ax=ax2)
    img3 = ax3.hist2d(in_map['S125'],in_map['HE_stoch_strong'],bins=100,cmap=plt.cm.jet,norm=matplotlib.colors.LogNorm())
    ax3.set_xlabel('S125')
    fig.colorbar(img3[3], ax=ax3)
    fig.tight_layout()
    fig.savefig(plot_name)
```

```python
plot_data(map1,'FeMaps.png')
plot_data(map2,'HeMaps.png')
plot_data(map3,'OMaps.png')
plot_data(map4,'PMaps.png')
```
