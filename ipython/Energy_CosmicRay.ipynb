```python
import numpy as np
```

```python
def get_data(file,allow_pickle=True):
    x = np.load(file,allow_pickle=allow_pickle)
    y = zip(*x)
    y1 = zip(y[0])
    y2 = zip(y[1])
    y1 = np.array(y1).flatten()
    y2 = np.array(y2).flatten()
    return y1,y2

def get_data_2(x):
    y = zip(*x)
    y1 = zip(y[0])
    y2 = zip(y[1])
    y1 = np.array(y1).flatten()
    y2 = np.array(y2).flatten()
    return [y1,y2]
```

```python
y1,y2 = get_data('/home/andres/CosmicRay.git/trunk/NN/Energy_output.npy')
x = np.load('/home/andres/CosmicRay.git/trunk/NN/Energy_split.npy',allow_pickle=True)
z1 = get_data_2(x[0])
z2 = get_data_2(x[1])
z3 = get_data_2(x[2])
z4 = get_data_2(x[3])
```

```python
import matplotlib.pylab as plt
import matplotlib
```

```python
plt.hist2d(z1[0],z1[1],bins=100,cmap=plt.cm.jet,norm=matplotlib.colors.LogNorm())
plt.plot(z1[0],z1[0],'r')
plt.xlabel('E_true')
plt.ylabel('E_rec')
plt.savefig('Helium_energy.png')
```

```python
plt.hist2d(z2[0],z2[1],bins=100,cmap=plt.cm.jet,norm=matplotlib.colors.LogNorm())
plt.plot(z2[0],z2[0],'r')
plt.xlabel('E_true')
plt.ylabel('E_rec')
plt.savefig('Proton_energy.png')
```

```python
plt.hist2d(z4[0],z4[1],bins=100,cmap=plt.cm.jet,norm=matplotlib.colors.LogNorm())
plt.plot(z4[0],z4[0],'r')
plt.xlabel('E_true')
plt.ylabel('E_rec')
plt.savefig('Oxygen_energy.png')
```
