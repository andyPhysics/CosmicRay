```python
x = np.load('/home/andres/CosmicRay.git/trunk/NN/All_split.npy',allow_pickle=True).item()
```

```python
def binned_values(bottom,top,data):
    binned_data = []
    for i,j in data:
        if i > bottom:
            if i <= top:
                binned_data.append(j)
    return np.array(binned_data)
```

```python
y = x['pred']
z1 = binned_values(7.4,7.5,y[0])
z2 = binned_values(7.4,7.5,y[1])
z3 = binned_values(7.4,7.5,y[2])
z4 = binned_values(7.4,7.5,y[3])
```

```python
import matplotlib.pylab as plt
import matplotlib
```

```python
z = binned_values(0,100,y[1])
plt.hist(z,bins=100)
plt.show()
```
