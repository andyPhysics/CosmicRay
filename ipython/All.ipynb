```python
def get_energy(data):
    energy = []
    for i,j in data:
        energy.append(i)
    return np.array(energy)
def binned_values(bottom,top,data):
    binned_data = []
    for i,j in data:
        if i > bottom:
            if i <= top:
                binned_data.append(j)
    return np.array(binned_data)
def all_binned_values(bottom,top,data):
    z1 = binned_values(bottom,top,data[0])
    z2 = binned_values(bottom,top,data[1])
    z3 = binned_values(bottom,top,data[2])
    z4 = binned_values(bottom,top,data[3])
    return z1,z2,z3,z4
def plot_stacked(bottom,top,title,output_file,data):
    import matplotlib.pylab as plt
    import matplotlib
    x1,x2,x3,x4 = all_binned_values(bottom,top,data)
    fig, axs = plt.subplots(4,sharex=True)
    fig.suptitle(title)
    axs[0].hist(x1, bins=100)
    axs[0].set_ylabel('Helium')
    axs[1].hist(x2,bins=100)
    axs[1].set_ylabel('Proton')
    axs[2].hist(x3,bins=100)
    axs[2].set_ylabel('Iron')
    axs[3].hist(x4,bins=100)
    axs[3].set_ylabel('Oxygen')
    plt.savefig(output_file)
```

```python
y = x['pred']
```

```python
plot_stacked(5.0,7.0,'Log(E) 5-7','All_plots_5_7.png',y)
```
