```python
import numpy as np
import pandas as pd
```

```python
x1 = pd.read_csv("proton_frac_values.csv",delimiter = ',')
x2 = pd.read_csv("helium_frac_values.csv",delimiter = ',')
x3 = pd.read_csv("oxygen_frac_values.csv",delimiter = ',')
x4 = pd.read_csv("iron_frac_values.csv",delimiter = ',')
```

```python
x1
```

```python
def combine_species(name):
    return pd.DataFrame({'log10(E/GeV)':x1['log10(E/GeV)'],'proton':x1[name],'helium':x2[name],'oxygen':x3[name],'iron':x4[name]})
```

```python
result1 = combine_species('H3A')
result2 = combine_species('H4A')
result3 = combine_species('GST')
result4 = combine_species('GST-3gen')
```

```python
result1.to_csv('H3A_values.csv',index = False,float_format='%.3f')
result2.to_csv('H4A_values.csv',index = False,float_format='%.3f')
result3.to_csv('GST_values.csv',index = False,float_format='%.3f')
result4.to_csv('GST-3gen_values.csv',index = False,float_format='%.3f')
```

```python
import seaborn
import matplotlib.pylab as plt
```

```python
def plotting_function(result,model,name):
    plt.plot(result['log10(E/GeV)'],result['proton'],label='proton',linestyle='--')
    plt.plot(result['log10(E/GeV)'],result['helium'],label='helium',linestyle='--')
    plt.plot(result['log10(E/GeV)'],result['oxygen'],label='oxygen',linestyle='--')
    plt.plot(result['log10(E/GeV)'],result['iron'],label='iron',linestyle='--')
    plt.grid(True,linestyle = 'dashed')
    plt.xlabel('log10(E/GeV)')
    plt.ylabel('Species Fraction')
    plt.title(model)
    plt.tight_layout()
    plt.legend()
    plt.show()
    #plt.savefig(name)
```

```python
plotting_function(result1,'H3A','H3A_plot.csv')
```

```python
plotting_function(result2,'H4A','H4A_plot.csv')
```

```python
plotting_function(result3,'GST','GST_plot.csv')
```

```python
plotting_function(result4,'GST-3gen','GST-3gen_plot.csv')
```
